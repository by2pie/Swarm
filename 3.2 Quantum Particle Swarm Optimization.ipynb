# 3.2 Implementation of Quantum Particle Swarm Optimization (QPSO)

In following notebook I would like to introduce idea and implementation of Quantum Particle Swarm Optimization

## 1. Main idea:

To make long story short we would like to use approach used previously in PSO, but instead of updating velocity and position use concept of wave function and its measurement in view of finding current position of particles. It shouldn't be complicated since it is possible to use lion part of previous code and only modify update equations.

## 2. Wave function and Schrödinger equation:

In the beginning I would like to make sure we are on the same page with understanding the concept of wave function. In quantum mechanics it is impossible to measure velocity and position at the same time (with arbitrary accuracy), thus instead of those two quantity we should assume existence of wave function, characteristic for every single particle. In something called position representation is simply given by function of position:

$$ \psi ({\bf x}) = f({\bf x}) $$

I would like to explain quantum mechanical kindergarden in lecture-like way, however today we should concentrate on QPSO and treat quanutm mechanics as a tool. Thus system evolution is govern by Schrödinger equation:

$$ i\hbar \frac{\partial \psi({\bf x})}{\partial t} = \hat{H} \psi({\bf x})$$

Which after transforming into stationary form (independent of time) yields:

 $$\hat{H} \psi({\bf x}) = E\psi({\bf x})$$

Where $\hat{H}$ is Hamiltionian of the system which is nothing else but Energy operator and E is scalar corresponding to energy for state $\psi({\bf x})$. In our case Hamiltonian will be simple, one particle Hamiltionian, namely it consist of two terms kinetic energy and potential energy.

$$\hat{H} = \frac{\hbar^2 \nabla^2}{2m} + V(x)$$

Since $\hat{p} = -i \hbar \nabla$ and potential energy is scalar fuction depending on current position. Finally linear equation for eigenstates of wavefunction is equal to:

<a id ='2.1'></a>
$$\bigg(\frac{\hbar^2 \nabla^2}{2m} + V(x)\bigg) \psi({\bf x}) = E\psi({\bf x}) \quad(2.1) $$

## 3. Probability density function:

By solving linear equation we can find eigenfunction $\psi({\bf x})$ for given particle, but what is really crucial is quantum mechanical feature connecting it with probability density function:

$$|\psi(x)|^2 = pdf(x)$$

By introducing potential into equation  <a href= 2.1>2.1</a>  we shall obtain quantum wells within probability of finding particle will be greatest. 

We can imagine now our particles are trapped (with non-zero tunneling probability in quantum wells) so it would be very nice to enforce convergance to our algorithm, it is done by addjusting following condition:

$$\lim_{t \to \infty} |\psi(x)|^2 = 0$$

It is done, by applying:

$$\int_{-\infty}^{|x_k|} |\psi(x)|^2_{k+1} dx >0.75 $$

Which is equivalent to:


$$\int_{0}^{|x_k|} |\psi(x)|^2_{k+1} dx >0.25$$

It means we are requiring at least 25% chances of particle to be within origin and actual position of k-th particle $x_k$

## 6. Implementation:

Main difference between PSO and QPSO lays within <b>particle</b> class, instead of update_velocity and update_position methods we shall define update function according to rules described above. It goes as follows:

```python
 def update(self, best_position, g, potential='delta'):
        """Method responsible for update, with implemented measurement operation
        Keyword arguments:
        best_position -- function defined as other method
        g -- constant responsible for covergance (>1)
        potential -- type of potential funciton default='delta'
        """
        phi = np.random.random(2)
        
        P = (phi[0] * self.best_position + phi[1] * best_position)/(phi[0]+phi[1])
        
        u = np.random.random(1)
        up = 0
        if potential == 'delta':
            up = 1/(2*g*log(2**0.5)) * abs(self.position - P)
        elif potential == 'harmonic':
            up = 1 / (0.47694*g) * (log(1/u))**0.5 * abs(self.position - P)

        if np.random.random(1) > 0.5:
            self.position = P + up
        else:
            self.position = P - up
```

## 7. Full code:

```python
__author__ = 'johnnyeleven'


import numpy as np
import matplotlib.pylab as plt
from math import *
from mpl_toolkits.mplot3d import Axes3D


class Particle(object):
    """ Class representing probe particle
    """

    def __init__(self, x):
        """Constructor for Particle
        Keyword arguments:
        x -- double[d] coordinates of particles in d dimensions
        v -- double[d] velocity vector of particles in d dimensions
        """
        self.position = x
        self.best_value = 9999
        self.best_position = self.position
        self.value = 0

    def get_value(self, fun):
        """Method getting value of function fun in current position
        Keyword arguments:
        fun -- function defined as other method
        """
        value = fun(self.position)
        if value < self.best_value:   # minimisation option
            self.best_value = value
            self.best_position = self.position
        self.value = value

    def update(self, best_position, g, potential='delta'):
        """Method responsible for update, with implemented measurement operation
        Keyword arguments:
        best_position -- function defined as other method
        g -- constant responsible for covergance (>1)
        potential -- type of potential funciton default='delta'
        """
        phi = np.random.random(2)

        P = (phi[0] * self.best_position + phi[1] * best_position)/(phi[0]+phi[1])

        u = np.random.random(1)
        up = 0
        if potential == 'delta':
            up = 1/(2*g*log(2**0.5)) * abs(self.position - P)
        elif potential == 'harmonic':
            up = 1 / (0.47694*g) * (log(1/u))**0.5 * abs(self.position - P)

        if np.random.random(1) > 0.5:
            self.position = P + up
        else:
            self.position = P - up



def function(position):
    """Definition of function to minimise
            position -- double[2] position of particle in d=2 dimensions
        Returns:
            value of function
    """
    x = position[0]
    y = position[1]
    return 20 + exp(1) - 20*exp(-0.2*((x**2+y**2)/2)**0.5)-exp(1/2*(cos(2*pi*x)+cos(2*pi*y)))


#initialisation of particles
N = 25 # number of particles
dt = 0.1 # time step
particles = []
x_begin = y_begin = -8
x_end = y_end = 8
best_value = 9999
best_position = np.array(2)
g = 1.7

for i in range(N):
    particles.append(Particle(2 * x_end * np.random.random(2) - x_end))  # case of random position, no velocity!


number_of_steps = 150

#plot initialisation
x = np.linspace(x_begin, x_end, 100)
y = np.linspace(y_begin, y_end, 100)
X, Y = np.meshgrid(x, y)
fig = plt.figure()
plt.ion()
plt.show()
Z = np.array([function((x, y)) for x, y in zip(np.ravel(X), np.ravel(Y))])  # calculating values of function
Z = Z.reshape(X.shape)

# main loop
for i in range(number_of_steps):
    values = []
    ax = fig.add_subplot(111, projection='3d')

    for particle in particles:  # calculation loop
        particle.get_value(function)
        values.append(particle.value)

    for particle in particles:  # update loop
        if particle.value < best_value:
            print('Bzzz! New minimum found: '+str(min(values)))
            print('min in: '+str(particle.position))
            best_value = particle.value
            best_position = particle.position

        particle.update(best_position, g)
        ax.scatter(particle.position[0], particle.position[1], particle.value, s=50, c='y')

    # plotting surface with transparency alpha = 0.2
    ax.plot_surface(X, Y, Z, alpha=0.2)
    plt.draw()
    plt.clf()
```
