# 3.1 Implementation of basic Particle Swarm Optimization (PSO)

In following notebook I would like to implement Particle Swarm Optimisation, a heuristic tool hoping to find extremum of any function in any dimension.

### 1. Theory and PSO blueprint

Firstly let mi start with presenting appropriate flow chart, describing flow of procedures:

```python
Image(url='http://openi.nlm.nih.gov/imgs/512/123/3231381/3231381_sensors-11-04721f3.png')
```

Where: <br>
-> $gBest$ is global best value of function $f({\bf x})$ among population <br>
-> $pBest$ is global person value of function $f({\bf x})$ for each particle


We ought to define update equations, namely velocity and position update for $^{i^{th}}$ particle are given by following formulas:

Velocity:
$$ {\bf v}^i(t_{i+1}) = c_1 {\bf v}^i(t_i) + c_2 r_1 \big(pBest^i(t_i) - {\bf x}^i(t_i)\big) + c_3 r_2 \big(gBest(t_i) - {\bf x}^i(t_i)\big) $$
Where <br>
$c_1$ - Coefficient responsible for intertia of particle (inertia) <br>
$c_2$ - Coefficient responsible for using personal knowladge (egoism) <br>
$c_3$ - Coefficient responsible for chasing global best position (obedience) <br>
$r_1, r_2 $ - Random numbers in interval $[0,1]$ <br>

As we can see from equation above we can divide update equation into three terms. Algorithm is very sensitive to coefficient ratio, for example maximizing the first one would result in unsteady particles oscillating around possible minimum. <br>

Assuming we already calculated $ {\bf v}^i(t_{i+1}) $, we should use our knowledge to calculate position update:


$$ {\bf x}^i(t_{i+1}) = {\bf x}^i(t_i) + {\bf v}^i(t_{i+1})  dt$$

This is rather straightforward Euler discretization technique.

## 3. Particle Class

As always, in programs and algorithms simulating motion of particles, it is useful to define class encapsulating all features and properties of object like particle. Because of complex comments I leave code without any further explanation:

```python
class Particle(object):
    """ Class representing probe particle
    """

    def __init__(self, x, v):
        """Constructor for Particle
        Keyword arguments:
        x -- double[d] coordinates of particles in d dimensions
        v -- double[d] velocity vector of particles in d dimensions
        """
        self.position = x
        self.velocity = v
        self.best_value = None
        self.best_position = self.position
        self.value = None

    def get_value(self, fun):
        """Method getting value of function fun in current position
        Keyword arguments:
        fun -- function defined as other method
        """
        value = fun(self.position)
        if value < self.best_value:   # minimisation option
            self.best_value = value
            self.best_position = self.position
        self.value = value

    def update_position(self, dt=1):
        """Method updating position of particle in current moment
        Keyword arguments:
        dt -- time step, default=1
        """
        self.position += np.array(self.velocity) * dt

    def update_velocity(self, global_best, coefficients):
        """Method updating velocity of particle in current moment
        Keyword arguments:
        global_best -- double[d] current best position of population, d-dimensions
        coefficients -- double[3] vector of three coefficients of simulation: 0-inertial 1-egoism 2-group terms
        """
        for i in range(len(global_best)):
            self.velocity[i] = self.velocity[i] * coefficients[0] + \
                coefficients[1] * np.random.random() * (self.best_position[i] - self.position[i]) + \
                coefficients[2] * np.random.random() * (global_best[i] - self.position[i])
```

## 3. Defining function:

Since we would like to visualize results it is obvious we should use function of two variable. In our case function, we are going to optimize(minimize) looks in the following way:

$$f(x,y) = 20 + e^{1} - 20e^{-0.2 \sqrt{\frac {x^2 + y^2}{2}}} - e^{\frac {cos(2\pi x) + cos(2\pi y)}{2}}$$

```python
Image(filen|ame='function.png', width=400)
```

```python
def function(position):
    """Definition of function to minimise
            position -- double[2] position of particle in d=2 dimensions
        Returns:
            value of function
    """
    x = position[0]
    y = position[1]
    return 20 + exp(1) - 20*exp(-0.2*((x**2+y**2)/2)**0.5)-exp(1/2*(cos(2*pi*x)+cos(2*pi*y)))
```

## 4. Initialization

Now the times has come to initialize variables. I would like to start with the case of zero initial velocity, but in general one can randomize it:

```python
#initialisation of particles
N = 10 # number of particles
dt = 0.1 # time step
particles = []
x_begin = y_begin = -8
x_end = y_end = 8
best_value = 99999

for i in range(N):
    particles.append(Particle(2 * x_end * np.random.random(2) - x_end, [0, 0]))  # case of random position, no velocity!

c = [0.6, 0.5, 0.9]   # [inertial term, egoism term, obedience term]
number_of_steps = 150

#plot initialisation
x = np.linspace(x_begin, x_end, 100)
y = np.linspace(y_begin, y_end, 100)
X, Y = np.meshgrid(x, y)
fig = plt.figure()
plt.ion()
plt.show()
Z = np.array([function((x, y)) for x, y in zip(np.ravel(X), np.ravel(Y))])  # calculating values of function
Z = Z.reshape(X.shape)
```

## 5. Main loop

Loop is calculating values for arbitrarily number of steps, inside main loop we can find two additional loops responsible for calculating new function values and updating process respectively. Finally we are plotting result using scatter and surface plots:

```python
# main loop
for i in range(number_of_steps):
    values = []
    ax = fig.add_subplot(111, projection='3d')

    for particle in particles:  # calculation loop
        particle.get_value(function)
        values.append(particle.value)

    for particle in particles:  # update loop
        if min(values) < best_value:
            print('Bzzz! New minimum found: '+str(min(values)))
        best_value = min(values)

        particle.update_velocity(particles[values.index(min(values))].position, c)
        particle.update_position(dt)
        ax.scatter(particle.position[0], particle.position[1], particle.value, s=50, c='y')

    # plotting surface with transparency alpha = 0.2
    ax.plot_surface(X, Y, Z, alpha=0.2)
    plt.draw()
    plt.clf()
```

```python
__author__ = 'johnnyeleven'


import numpy as np
import matplotlib.pylab as plt
from math import *
from mpl_toolkits.mplot3d import Axes3D


class Particle(object):
    """ Class representing probe particle
    """

    def __init__(self, x, v):
        """Constructor for Particle
        Keyword arguments:
        x -- double[d] coordinates of particles in d dimensions
        v -- double[d] velocity vector of particles in d dimensions
        """
        self.position = x
        self.velocity = v
        self.best_value = 0
        self.best_position = self.position
        self.value = 0

    def get_value(self, fun):
        """Method getting value of function fun in current position
        Keyword arguments:
        fun -- function defined as other method
        """
        value = fun(self.position)
        if value < self.best_value:   # minimisation option
            self.best_value = value
            self.best_position = self.position
        self.value = value

    def update_position(self, dt=1):
        """Method updating position of particle in current moment
        Keyword arguments:
        dt -- time step, default=1
        """
        self.position += np.array(self.velocity) * dt

    def update_velocity(self, global_best, coefficients):
        """Method updating velocity of particle in current moment
        Keyword arguments:
        global_best -- double[d] current best position of population, d-dimensions
        coefficients -- double[3] vector of three coefficients of simulation: 0-inertial 1-egoism 2-group terms
        """
        for i in range(len(global_best)):
            self.velocity[i] = self.velocity[i] * coefficients[0] + \
                coefficients[1] * np.random.random() * (self.best_position[i] - self.position[i]) + \
                coefficients[2] * np.random.random() * (global_best[i] - self.position[i])


def function(position):
    """Definition of function to minimise
            position -- double[2] position of particle in d=2 dimensions
        Returns:
            value of function
    """
    x = position[0]
    y = position[1]
    return 20 + exp(1) - 20*exp(-0.2*((x**2+y**2)/2)**0.5)-exp(1/2*(cos(2*pi*x)+cos(2*pi*y)))


#initialisation of particles
N = 10 # number of particles
dt = 0.1 # time step
particles = []
x_begin = y_begin = -8
x_end = y_end = 8
best_value = 0

for i in range(N):
    particles.append(Particle(2 * x_end * np.random.random(2) - x_end, [0, 0]))  # case of random position, no velocity!

c = [0.6, 0.5, 0.9]   # [inertial term, egoism term, obedience term]
number_of_steps = 10

#plot initialisation
x = np.linspace(x_begin, x_end, 100)
y = np.linspace(y_begin, y_end, 100)
X, Y = np.meshgrid(x, y)
fig = plt.figure()
plt.ion()
plt.show()
Z = np.array([function((x, y)) for x, y in zip(np.ravel(X), np.ravel(Y))])  # calculating values of function
Z = Z.reshape(X.shape)

# main loop
for i in range(number_of_steps):
    values = []
    ax = fig.add_subplot(111, projection='3d')

    for particle in particles:  # calculation loop
        particle.get_value(function)
        values.append(particle.value)

    for particle in particles:  # update loop
        if min(values) < best_value:
            print('Bzzz! New minimum found: '+str(min(values)))
        best_value = min(values)

        particle.update_velocity(particles[values.index(min(values))].position, c)
        particle.update_position(dt)
        ax.scatter(particle.position[0], particle.position[1], particle.value, s=50, c='y')

    # plotting surface with transparency alpha = 0.2
    ax.plot_surface(X, Y, Z, alpha=0.2)
    plt.draw()
    plt.clf()
```
